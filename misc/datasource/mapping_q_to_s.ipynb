```python
# import libraries needed
import IPython
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load mapping dictionary file
df_map = pd.read_csv('mapping_dict_q_s.csv')
print(df_map.shape)
df_map.head()
```

```python
# convert it into a dictionary
mapdict = {}
mapsmid = {}
for index, row in df_map.iterrows():
    mapdict[row[0]] = row[2]
    mapsmid[row[2]] = row[1]

print(len(mapdict))
print(len(mapsmid))
```

```python
# load data file containing questions
df = pd.read_csv('dummy_top5_symptoms.csv')
df.drop(['questionnaire', 'Unnamed: 0'], axis=1, inplace=True)
print(df.shape)
df.head()
```

```python
# apply mapping dictionary to convert questions to symptoms + symptom IDs
df['symptom'] = df['question']
df['symptom'].replace(mapdict, inplace=True)
df['symp_id'] = df['symptom']
df['symp_id'].replace(mapsmid, inplace=True)
df.head()
```

```python
# output to file
df.to_csv('dummy_symptoms_mapped.csv')
```
